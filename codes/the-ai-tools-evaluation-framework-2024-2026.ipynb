```python
# ══════════════════════════════════════════════════════
# CELL 3 — MATHEMATICAL ANALYSIS ENGINE
# ══════════════════════════════════════════════════════

def composite_score(model_data, weights=None):
    """
    Weighted composite score for LLMs.
    Score = sum(w_i * metric_i) / sum(w_i)
    """
    metrics = ["mmlu", "humaneval", "math", "reasoning", "coding", "multilingual"]
    if weights is None:
        weights = [0.20, 0.20, 0.15, 0.20, 0.15, 0.10]
    score = sum(model_data[m] * w for m, w in zip(metrics, weights))
    return round(score, 2)

def cost_efficiency_score(model_data, category="llm"):
    """
    Cost-Performance Ratio:
    CPS = Performance / log(Cost + 1)
    Higher = better value for money
    """
    if category == "llm":
        perf = composite_score(model_data)
        cost = model_data["cost_per_1m"] + 0.01
    elif category == "image":
        perf = np.mean([model_data["fidelity"], model_data["creativity"],
                        model_data["prompt_following"]])
        cost = (model_data["cost_per_img"] * 1000) + 0.01
    else:
        perf = np.mean([model_data["completion_acc"], model_data["bug_detection"]])
        cost = model_data["cost_monthly"] + 0.01
    return round(perf / np.log(cost + 1), 2)

def scalability_index(params_b, mmlu_score):
    """
    Scalability: measures how efficiently a model uses its parameters.
    SI = MMLU / log2(params)
    Higher = more efficient use of scale
    """
    return round(mmlu_score / np.log2(params_b + 1), 2)

def elo_rating_simulation(models_data, n_battles=10000):
    """
    Simulate ELO tournament between models.
    P(A beats B) = 1 / (1 + 10^((Rb - Ra) / 400))
    """
    names = list(models_data.keys())
    ratings = {n: 1200 for n in names}
    scores = {n: composite_score(models_data[n]) for n in names}

    for _ in range(n_battles):
        a, b = np.random.choice(names, 2, replace=False)
        expected_a = 1 / (1 + 10**((ratings[b] - ratings[a]) / 400))
        actual_a = 1.0 if scores[a] > scores[b] + np.random.normal(0, 3) else 0.0
        k = 32
        ratings[a] += k * (actual_a - expected_a)
        ratings[b] += k * ((1 - actual_a) - (1 - expected_a))

    return {n: round(ratings[n]) for n in names}

# ── Compute all metrics ────────────────────────────────
print("Computing mathematical metrics...")

llm_composite = {m: composite_score(d) for m, d in LLM_DATA.items()}
llm_cps       = {m: cost_efficiency_score(d, "llm") for m, d in LLM_DATA.items()}
llm_si        = {m: scalability_index(d["params_b"], d["mmlu"]) for m, d in LLM_DATA.items()}
llm_elo       = elo_rating_simulation(LLM_DATA)

img_composite = {m: round(np.mean([d["fidelity"],d["creativity"],
                                    d["prompt_following"],d["consistency"]]),1)
                 for m, d in IMAGE_DATA.items()}
img_cps       = {m: cost_efficiency_score(d, "image") for m, d in IMAGE_DATA.items()}

code_composite = {m: round(np.mean([d["completion_acc"],d["bug_detection"],
                                     d["refactoring"],d["test_generation"]]),1)
                  for m, d in CODE_DATA.items()}

print("\nLLM Composite Scores:")
for m, s in sorted(llm_composite.items(), key=lambda x: -x[1]):
    print(f"  {m:<25} {s:>6.2f}  ELO={llm_elo[m]}  CPS={llm_cps[m]}")
print("\nImage Model Scores:")
for m, s in sorted(img_composite.items(), key=lambda x: -x[1]):
    print(f"  {m:<25} {s:>6.1f}  CPS={img_cps[m]}")
print("\nCode Tool Scores:")
for m, s in sorted(code_composite.items(), key=lambda x: -x[1]):
    print(f"  {m:<25} {s:>6.1f}")
```

## Mathematical Metrics Used

| Metric | Formula | Measures |
|--------|---------|----------|
| Composite Score | Weighted avg of 6 benchmarks | Overall capability |
| ELO Rating | Tournament simulation (10k battles) | Competitive ranking |
| Cost Efficiency | Performance / log(Cost) | Value for money |
| Scalability Index | MMLU / log2(Params) | Parameter efficiency |

## Key Findings

The best model is not always the most expensive.
The fastest model is not always the most accurate.
The safest model is not always the most capable.

This framework measures all three simultaneously.
